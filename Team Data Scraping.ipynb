```python
import requests
```

# Shot Info.

```python
url = "http://www.nhl.com/stats/rest/team?isAggregate=false&reportType=shooting&isGame=false&reportName=teamsummaryshooting&sort=[{%22property%22:%22shotAttempts%22,%22direction%22:%22DESC%22}]&cayenneExp=gameTypeId=2%20and%20seasonId%3E=20172018%20and%20seasonId%3C=20172018"
data = requests.get(url).json()
```

```python
out = {team["teamFullName"] : {"gp" : team["gamesPlayed"], "bsA" : team["shotAttemptsAgainst"] - team["unblockedShotAttempts"]} for team in data["data"]}
```

# Standard Info.

```python
url = "http://www.nhl.com/stats/rest/team?isAggregate=false&reportType=basic&isGame=false&reportName=teamsummary&sort=[{%22property%22:%22points%22,%22direction%22:%22DESC%22},{%22property%22:%22wins%22,%22direction%22:%22DESC%22}]&cayenneExp=gameTypeId=2%20and%20seasonId%3E=20172018%20and%20seasonId%3C=20172018"
data = requests.get(url).json()["data"]
```

```python
for team in data:
    out[team["teamFullName"]].update({
        "ga" : team["goalsAgainst"],
        "gf" : team["goalsFor"],
        "saa" : team["shotsAgainstPerGame"] * out[team["teamFullName"]]["gp"],
        "saf" : team["shotsForPerGame"] * out[team["teamFullName"]]["gp"],
        "totLosses" : team["losses"] + team["otLosses"],
        "sh%" : team["goalsAgainst"]/(team["shotsAgainstPerGame"] * out[team["teamFullName"]]["gp"]),
    })
```
